```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('/Users\LAPTOP EXPERT\Downloads/')
```

```python
df.head()
```

```python
df.info()
```

```python
# we want only last two columns for prediction
x = df.iloc[:,[3,4]].values
```

```python
# finding optimum value of k by elbow method
k_inertia = []
for i in range(1,11):
    model = KMeans(n_clusters = i,init='k-means++',random_state = 0)
    model.fit(x)
    k_inertia.append(model.inertia_)
```

```python
plt.plot(range(1,11),k_inertia)
plt.xlabel('Value of k')
plt.ylabel('Value of inertia')
plt.show()
```

```python
# we come to know that at k = 5 inertia heavily change after that inertia is uniform and no elbow point
model = KMeans(n_clusters=5,init='k-means++',random_state=2)
model.fit(x)
predicted_y_clusters = model.predict(x)
```

```python
centroids = model.cluster_centers_
```

```python
centroids
```

```python
plt.figure(figsize=(8,6))
plt.scatter(centroids[:,0],centroids[:,1],s=220,c='b')
plt.scatter(x[:,0],x[:,1],c = predicted_y_clusters)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x = centroids[:,0],y = centroids[:,1],data=centroids,s =120)
sns.scatterplot(x = x[:,0],y=x[:,1],data=x,hue=predicted_y_clusters,s=50,palette='summer')
plt.xlabel('Annual Income',fontdict={'size':15,'color':'b'})
plt.ylabel('Spending Score',fontdict={'size':15,'color':'b'}) 
plt.show()
```
